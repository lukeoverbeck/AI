**Problem 3 (10 points)**

Write a **DFS** algorithm which when given a **start node** and a **goal node**, it returns **SUCCESS** if the goal is reachable and **FAILURE** otherwise. Your algorithm should test both graphs shown below and return **SUCCESS** for the first graph and **FAILURE** for the second one.

```python
from collections import deque

# Start the graph search from root using a doubly-ended queue
# both these combinations work for DFS: append/pop and appendleft/popleft

def DFS_maze(graph, start, end):

    # deque is list-like container with fast appends and pops on either end
    visited, queue = set(), deque([start])
    visited.add(start)

    # Keep exploring as long as the queue is not empty
    while queue:

        # Dequeue a vertex from queue
        vertex = queue.pop()

        # Return success if goal reached
        if vertex == end:
          return f"Success! Ending point {end} reached from starting point {start}"

        # If not visited, mark the node as visited and enqueue it
        # Use reversed() to enforce left-to-right ordering

        for neighbor in reversed(graph[vertex]):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    # Return fail if goal not reached
    return f"Fail! Ending point {end} not reached from starting point {start}"
```
